```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
def global_alignment(seq1, seq2):
    """
    Perform global alignment using Needleman‚ÄìWunsch algorithm.
    """
    print("üîπ Performing Global Alignment (Needleman‚ÄìWunsch)...\n")
    alignments = pairwise2.align.globalxx(seq1, seq2)

    for alignment in alignments[:1]:
        print(format_alignment(*alignment))
        print("Alignment Score:", alignment[2])
    return alignments[0]

def local_alignment(seq1, seq2):
    """
    Perform local alignment using Smith‚ÄìWaterman algorithm.
    """
    print("üîπ Performing Local Alignment (Smith‚ÄìWaterman)...\n")
    alignments = pairwise2.align.localxx(seq1, seq2)

    for alignment in alignments[:1]:
        print(format_alignment(*alignment))
        print("Alignment Score:", alignment[2])
    return alignments[0]
```

```python

seq1 = "ATGCGTACGT"
seq2 = "ATGCGACGTAGC"

print("Sequence 1:", seq1)
print("Sequence 2:", seq2)
```

```python

global_result = global_alignment(seq1, seq2)

local_result = local_alignment(seq1, seq2)
```

```python

print("üîπ Custom Scoring Example (DNA)...\n")

alignments = pairwise2.align.globalms(seq1, seq2, 2, -1, -0.5, -0.1)
for a in alignments[:1]:
    print(format_alignment(*a))
    print("Alignment Score:", a[2])
```

```python
import matplotlib.pyplot as plt

def visualize_alignment(seq1, seq2, alignment):
    aligned_seq1, aligned_seq2, score, start, end = alignment
    matches = [1 if a == b else 0 for a, b in zip(aligned_seq1, aligned_seq2)]

    plt.figure(figsize=(10, 2))
    plt.bar(range(len(matches)), matches, color='green')
    plt.title(f"Alignment Match Visualization (Score: {score})")
    plt.xlabel("Alignment Position")
    plt.ylabel("Match (1=True, 0=False)")
    plt.show()


visualize_alignment(seq1, seq2, global_result)
```
